# Learning MongoDB 

**Connecting MongoDB with Python and querying interesting results using BestBuy sample data**

```python
import pandas as pd
from pymongo import MongoClient


client = MongoClient("mongodb://localhost:27017/")ß

db = client["bestbuy_db"]  

collection = db["search_queries"] 

file_path = "bestbuy.csv" 
df = pd.read_csv(file_path)

# setup schema
df = df.rename(columns={
    "User": "user",
    "sku": "sku",
    "category": "category",
    "query": "query",
    "click_time": "click_time",
    "query_time": "query_time"
})

# Convert DataFrame to dictionary format for MongoDB insertion
data = df.to_dict(orient="records")

# Insert data into MongoDB
collection.insert_many(data)
```

```python
record_count = collection.count_documents({})

print(f"Total number of records in the database: {record_count}")
```

```python
pipeline = [
    {
        "$group": {
            "_id": "$sku",  # Group by 'sku'
            "count": {"$sum": 1}  # Count the occurrences
        }
    },
    {
        "$sort": {"count": -1}  # Sort by count in descending order
    },
    {
        "$limit": 5 
    }
]

top_skus = collection.aggregate(pipeline)

for sku in top_skus:
    print(f"sku: {sku['_id']}, Count: {sku['count']}")
```
